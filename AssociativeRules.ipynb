# Apriori

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_csv('Data/creditcard.csv')
data.head()
```

```python
data = data.drop(['Time', 'Amount'], axis=1)
```

```python
# Fradulent Transactions
data_fraud = (data[data['Class'] == 1]).reset_index(drop=True)
		#.groupby(['InvoiceNo', 'Description'])['Quantity']
		#.sum().unstack().reset_index().fillna(0))
		#.set_index('InvoiceNo'))

# Non Fradulent Transactions
data_regular = (data[data['Class'] == 0]).reset_index(drop=True)
		#.groupby(['InvoiceNo', 'Description'])['Quantity']
		#.sum().unstack().reset_index().fillna(0))
		#.set_index('InvoiceNo'))

#data_fraud.head()
#data_regular.head()
```

```python
data_fraud = data_fraud.drop(['Class'], axis=1)
data_regular = data_regular.drop(['Class'], axis=1)
```

```python
data_regular
```

```python
# Defining the hot encoding function to make the data suitable 
# for the concerned libraries
def hot_encode(x):
    if(x<= 0):
        return False
    if(x>= 0):
        return True
  
# Encoding the datasets
fraud_encoded = data_fraud.applymap(hot_encode)
regular_encoded = data_regular.applymap(hot_encode)
```

```python
fraud_encoded = fraud_encoded.fillna(0)
regular_encoded = regular_encoded.fillna(0)
```

```python
# Building the model
frq_items = apriori(fraud_encoded, min_support = 0.8, use_colnames = True)

# Collecting the inferred rules in a dataframe
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
print(rules.head())
```

```python
pd.DataFrame(rules)
```

```python
# Building the model
frq_items_regular = apriori(regular_encoded, min_support = 0.3, use_colnames = True)

# Collecting the inferred rules in a dataframe
rules_regular = association_rules(frq_items_regular, metric ="lift", min_threshold = 1)
rules_regular = rules_regular.sort_values(['confidence', 'lift'], ascending =[False, False])
print(rules_regular.head())
```

```python
pd.DataFrame(rules_regular)
```

```python
data_encoded = data.applymap(hot_encode)
data_encoded = data_encoded.fillna(0)
```

```python
list = []
for index, row in data_encoded.iterrows():
    if row.Class:
        list.append(False)
    else:
        list.append(True)

#print(list)
data_encoded['Not class'] = list
```

```python
# Building the model
frq_items_data = apriori(data_encoded, min_support = 0.2, use_colnames = True)

# Collecting the inferred rules in a dataframe
rules_data = association_rules(frq_items_data, metric ="lift", min_threshold = 1)
rules_data = rules_data.sort_values(['confidence', 'lift'], ascending =[False, False])
print(rules_data.head())
```

```python
rules_data_df = pd.DataFrame(rules_data)
```

```python
rules_data_df.consequents.unique()
```

```python
for index, row in rules_data_df.iterrows():
    # print(row.consequents)
    if 'Class' in row.consequents or 'Not class' in row.consequents and row.support > 0.5:
        print(row['antecedents'], row['support'])
```
